## Sample visualizations

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import datetime
warnings.simplefilter("ignore")
```

```python
df=pd.DataFrame({
    'x': [datetime.date(2019, 1, 1) + datetime.timedelta(days=10*i) for i in range(10)],
    'Trump': [0.5 - 0.05*x + 0.5*np.random.rand() for x in range(10)], 
    'Warren': [0.05*x + 0.5*np.random.rand() for x in range(10)], 
    'Biden':  [-0.1* x + 0.5*np.random.rand() for x in range(10)] 
})

plt.plot( 'x', 'Trump', data=df, marker='o', markerfacecolor='orange', markersize=10, color='goldenrod', linewidth=3, linestyle='dashed')
plt.plot( 'x', 'Biden', data=df, marker='o', markerfacecolor='red', markersize=10, color='orangered', linewidth=3, linestyle='dashed')
plt.plot( 'x', 'Warren', data=df, marker='o', markerfacecolor='blue', markersize=10, color='blue', linewidth=3, linestyle='dashed')

plt.xlabel('Time')
plt.ylabel('Sentiment Score')
plt.title('Sentiment of Tweets About Each Candidate')
plt.gcf().autofmt_xdate()

plt.legend()
    
plt.savefig('sample-plot.jpg')
```
